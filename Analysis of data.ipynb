# Missing Values 

```python
# Here we will check the percentage of Nan values in each feature
# get list of feature with missing values
feature_with_nan=[features for features in dataset.columns if dataset[features].isnull().sum()>1]
# print the feature name and percentage of missing values
for feature in feature_with_nan:
    print(feature,np.round(dataset[feature].isnull().mean(),4))
```

# Since they are many missing values we need to find the relationship between missing values and Sales price
Let's plot some diagram for this relationship


```python
for feature in feature_with_nan:
    data=dataset.copy()
    # Let make a variable that indicated 1 if observation are missing
    data[feature]=np.where(data[feature].isnull(),1,0)
# Let's calculate the mean salesprice where the information is missing 
   
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

Here with relation between the missing values and dependent variable is clearly visible.So we need to replace these nan values with something meaningful which we will do in the Feature Engineering section.

# Numerica Variables

```python
# List of numerical variables
# 'O' Here means object notation
numerical_features=[features for features in dataset.columns if dataset[features].dtypes !='O']
print(f'No. of numerical features are {len(numerical_features)}')

# Visualise the numerical variables
dataset[numerical_features].head()
```

# Temporal Variables(Eg. Datatime variables)

From the Dataset we have 4 year variables.We have extract information from the datetime variables like no of years or no of days.One example in this specific scenario can be difference in years between the year the house was built and the year the house was sold.We will be performing the analysis in the Feature Engineering

```python
# List of variables that contain year information
year_feature=[feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
# lets analyse the Temporal Datetime Variables
# We will check whether there is a relation between year the house is sold 
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price Vs yearSold')
```

```python
# here we will compare the difference between All years Feature with Salesprice
for feature in year_feature:
    if feature !='YrSold':
        data=dataset.copy()
        # We will take differecne between year sold and year variable
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.show()
```

## Discrete Feature

```python
# Numerical variable are usually 2 type
# 1. Continuous and 2. Discrete

discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print('Discrete Variable Count ',len(discrete_feature))
```

```python
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
# Let's find the relationship between them and Sale Price
for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

## Continuous Feature

```python
continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature+year_feature+['Id']]
print(f'Total no. of continuous feature ',len(continuous_feature))
```

```python
# Let's analyse the continuous values by creating a histogram
for feature in continuous_feature:
    data=dataset.copy()
    data[feature].hist()
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

```python
# using Logarithmic distribution
for feature in continuous_feature:
    data=dataset.copy()
    # becuase log0 is -inf
    if 0 in data[feature].unique() or feature =='SalePrice':
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(f'{feature}')
        plt.ylabel('SalePrice')
        plt.show()
```

# Outliers
For Outliers we use box plot

```python
for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique() or feature =='SalePrice':
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.show()
    
```
